```python
import pandas as pd
```

```python
# importar arquivo
tabela = pd.read_csv('./municipios.csv')
```

```python
tabela.head()
```

```python
selecao_capital = tabela.query('is_capital == True')
selecao_capital.reset_index(drop=True)
```

```python
mask = selecao_capital.sort_values(by='uf_code').reset_index(drop=True)
qtd_dados = len(mask.index)
json_todas_cidades = {}
for index in range(0,qtd_dados):
    cidade_nova = {
        'municipio': mask.iloc[index]['municipio'],
        'uf': mask.iloc[index]['uf'],
        'uf_code': mask.iloc[index]['uf_code'],
    }
    #print(cidade_nova)
    json_todas_cidades.update(cidade_nova)
print(json_todas_cidades)
```

```python
# montando dados via pandas
import json
selecao_capital = tabela.query('is_capital == True')
mask = selecao_capital.sort_values(by='uf_code').reset_index(drop=True)
tabela_json = mask.to_json()

# criando novo JSON
qtd_dados = len(mask.index)
dict_todas_cidades = {}
for index in range(0,qtd_dados):
    cidade_nova = {
        'municipio': str(mask.iloc[index]['municipio']),
        'uf': str(mask.iloc[index]['uf']),
        'uf_code': str(mask.iloc[index]['uf_code']),
    }
    #print(cidade_nova)
    dict_todas_cidades |= cidade_nova

# coverter em JSON
json_todas_cidades = json.dumps(dict_todas_cidades)
print(json_todas_cidades)
```
